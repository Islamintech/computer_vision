```python
from ddgs import DDGS
import requests
import os
import shutil

# 1. Define your new food type
FOOD_NAME = "pilaf"
SEARCH_QUERY = "uzbek pilaf plov food"
TARGET_IMAGES = 100

# 2. Setup local and Google Drive directories
local_dir = f"{FOOD_NAME}_images"
drive_target_dir = f"/content/drive/MyDrive/food101_dataset/{FOOD_NAME}"

shutil.rmtree(local_dir, ignore_errors=True)
os.makedirs(local_dir, exist_ok=True)
os.makedirs(drive_target_dir, exist_ok=True)

# 3. Download the images locally
saved = 0
print(f"Starting download for {FOOD_NAME}...")
with DDGS() as ddgs:
    results = ddgs.images(SEARCH_QUERY, max_results=300)
    for r in results:
        if saved >= TARGET_IMAGES:
            break
        try:
            img = requests.get(r["image"], timeout=10)
            if img.status_code == 200:
                # Save locally with a clean filename
                filename = f"{saved}.jpg"
                local_path = os.path.join(local_dir, filename)
                with open(local_path, "wb") as f:
                    f.write(img.content)

                # Move directly to Google Drive
                shutil.move(local_path, os.path.join(drive_target_dir, filename))
                saved += 1
                print(f"{saved} ok -> moved to Drive")
        except Exception:
            pass

print(f"\nFinished! Total {FOOD_NAME} images successfully saved to Drive: {saved}")
```
