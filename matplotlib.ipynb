```python
import matplotlib.pyplot as plt
import numpy as np
```

# **Basic Lines with Matplotlib**

```python
x = np.array([1,3,4,6])
y = np.array([20,30,40,6])

plt.plot(x,y)
plt.title("Basic Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

# **Styled Lines with Matplotlib**

```python
from matplotlib.lines import lineStyles
x = np.array([1,2,3,5])
y = np.array([3,8,1,5])

plt.plot(y, linestyle='dotted', color='green', marker='o')
plt.title("Styled Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

# **Multiple Lines with Matplotlib**

```python
x = np.array([1,2,3,4])
y1 = np.array([1,6,1,9])
y2 = np.array([2,7,2,10])

plt.plot(x, y1, label='Line 1', color='red', marker='s')
plt.plot(x, y2, label='Line 2', color='blue', linestyle='--', marker='o')
plt.title("Multiple Lines Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend() # qaysi rang qanday marker
plt.show()
```

# **Bar Chart with Numpy and Matplotlib**

```python
x = np.array(["Salary", "Expense", "Charaty", "Food"])
y = np.array([4,9,1,7])

plt.bar(x,y, color='green')
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

# **Scatter Plot with Matplotlib**

```python
x = np.array([1,2,3,4])
y = np.array([1,6,1,9])

plt.scatter(x,y, color='purple', s=100)
plt.title("Scatter Plot Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

# **Histogram Lines with Matplotlib**

```python
data = np.array([3,4,5,6,8,8,8,9,10])

plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Values")
plt.ylabel("Count")
plt.show()
```

# **Subplots(Multiple Graphs together)**

```python
x = np.array([1,2,3,4])
y1 = np.array([1,6,1,9])
y2 = np.array([2,7,2,10])

plt.figure(figsize=(8,4))   # 8 => with (wide) | 4 => height (tall)

plt.subplot(1,2,1)
plt.plot(x, y1, color='red')
plt.title('Line Plot - 1')

plt.subplot(1,2,2)
plt.bar(x, y2, color='blue')
plt.title('Bar Plot - 2')
plt.tight_layout()

plt.show()
```

```python
N_points = 10_000

# Generate two normal distributions
dataset1 = np.random.normal(0, 1, size=N_points)
dataset2 = np.random.normal(1, 2, size=N_points)

# Use a constant bin width to make the two histograms easier to compare visually
bin_width = 0.25
bins = np.arange(np.min([dataset1, dataset2]),
                    np.max([dataset1, dataset2]) + bin_width, bin_width)

fig, ax = plt.subplots()

# Plot the first histogram
ax.hist(dataset1, bins=bins, label="Dataset 1")

# Plot the second histogram
# (notice the negative weights, which flip the histogram upside down)
ax.hist(dataset2, weights=-np.ones_like(dataset2), bins=bins, label="Dataset 2")
ax.axhline(0, color="k")
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


def midpoints(x):
    sl = ()
    for _ in range(x.ndim):
        x = (x[sl + np.index_exp[:-1]] + x[sl + np.index_exp[1:]]) / 2.0
        sl += np.index_exp[:]
    return x

# prepare some coordinates, and attach rgb values to each
r, g, b = np.indices((17, 17, 17)) / 16.0
rc = midpoints(r)
gc = midpoints(g)
bc = midpoints(b)

# define a sphere about [0.5, 0.5, 0.5]
sphere = (rc - 0.5)**2 + (gc - 0.5)**2 + (bc - 0.5)**2 < 0.5**2

# combine the color components
colors = np.zeros(sphere.shape + (3,))
colors[..., 0] = rc
colors[..., 1] = gc
colors[..., 2] = bc

# and plot everything
ax = plt.figure().add_subplot(projection='3d')
ax.voxels(r, g, b, sphere,
          facecolors=colors,
          edgecolors=np.clip(2*colors - 0.5, 0, 1),  # brighter
          linewidth=0.5)
ax.set(xlabel='r', ylabel='g', zlabel='b')
ax.set_aspect('equal')

plt.show()
```
